```python
import pandas as pd 
import numpy as np
import statsmodels.api as sm

from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split

from matplotlib import pyplot as plt
```

```python
file_path = "/Users/owner/Analysis_Projects/sportsref_download_2021.csv"
file_path_two = "/Users/owner/Analysis_Projects/sportsref_download_2022.csv"

baseball_2021_df = pd.read_csv(file_path)
baseball_2022_df = pd.read_csv(file_path_two)

baseball_2021_df
```

```python
baseball_2022_df
```

# Here are Linear and Logistic Regressions for 2021 Baseball Total Runs 

```python
# Here, we will do linear regrerssion for perdicted runs for the seasons in 20201 and 2022
# The regression coefficients are 
# 
# First, here is 2021. 

runs_df_2021 = baseball_2021_df[["R", "BB+HBP",'2B', '3B', "HR", "SB", "CS"]]
runs_df_2021 = runs_df_2021.drop(runs_df_2021.index[-1])

# with sklearn 
regr = linear_model.LinearRegression()
regr.fit(runs_df_2021[[ "BB+HBP",'2B', '3B', "HR", "SB", "CS"]],runs_df_2021.R)


```

```python
regr.coef_

```

```python
x = runs_df_2021.drop(columns=["R"])
y = runs_df_2021["R"]

r_squared = regr.score(x, y)
print(r_squared)
```

```python
print(f"Intercept: {regr.intercept_} \n")
print(f"Coefficients: {regr.coef_}\n\n")

# This is the statistic model 

x = sm.add_constant(x) # adding a constant 
model = sm.OLS(y,x).fit()
predictions = model.predict(x)

print_model = model.summary()
print(f"This is the results for the runs predicted \n\n {print_model}")


```

```python
# Import the Confusion Matrix 

X = runs_df_2021.drop(columns=["R"])
y = runs_df_2021["R"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# len(X_train)
# len(X_test)

X_train
```

```python
# Fit the Model (Logistic Regression)

model = LogisticRegression(solver='lbfgs', max_iter=1000000000)
model.fit(X_train,y_train)
```

```python
#Make Predictions of The Logistic Regression Model and Make DataFrame between actual and predicted results
y_pred = model.predict(X_test)
results = pd.DataFrame({"Predicted":y_pred, "Actual":y_test}).reset_index(drop=True)
results
```

```python
# Accuracy Score (Logistic Regression)
score_one = model.score(X_test, y_test)
print(f"The Score For Logistic Regression: {score_one}")
```

```python
# Make Predictions for the Linear Regression Model and Dataframe comparing actual and predicted values 
model2 = LinearRegression()
model2.fit(X_train, y_train)
y_pred = model2.predict(X_test)
results = pd.DataFrame({"Predicted":y_pred, "Actual":y_test}).reset_index(drop=True)
results
```

```python
# Accuracy Score (Linear Regression)
from sklearn.metrics import accuracy_score
print(f"The Accuracy For Linear Regression :{accuracy_score(y_test, y_pred)}")
```
